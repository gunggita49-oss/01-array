```python
import numpy as np
import matplotlib.pyplot as plt

def program_nilai():
    print("=== INPUT 10 NILAI MAHASISWA ===")
    data_nilai = []
    # Kerangka awal program
    # 1. Input 10 nilai (O(n))
    for i in range(10):
        while True:
            try:
                n = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
                if 0 <= n <= 100:
                    data_nilai.append(n)
                    break
                else:
                    print("⚠️ Nilai harus 0-100.")
            except ValueError:
                print("⚠️ Masukkan angka!")
```
